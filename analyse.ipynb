# 3 - Traitement des données

```python
data_csv
```

```python
data_csv.dtypes
```

```python
data_csv.isna().sum()
```

```python
numero=data_csv.select_dtypes(exclude=['object']).columns
numero
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy ='mean')
imputer = imputer.fit(data_csv[numero]) 
data_csv[numero] = imputer.transform(data_csv[numero]).round(1)
data_csv.head(9)
```

```python
data_csv.isna().sum()
```

```python
obj=data_csv.select_dtypes(include=['object'])
obj
```

```python
obj=obj.iloc[:,1:4]
obj
```

```python
for i in obj.columns:
    le = preprocessing.LabelEncoder()
    le.fit(data_csv[i])
    data_csv[i] = le.transform(data_csv[i])

data_csv
```

```python
X=data_csv.iloc[:,1:-1]
y=data_csv['Solvable']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
len(X_train)
```

```python
len(X_test)
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X_train)
scaled_data
```

```python
scaler = StandardScaler()
scaled_d = scaler.fit_transform(X_test)
scaled_d
```
